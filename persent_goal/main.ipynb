```python
d = 30
j = 0
cumulasub_15sec = 0
cumulawatt_now = 0


time_a = pd.date_range('2016-10-01 00:00:00', periods=5760*d, freq='15S')
# print time_a
time_b = pd.date_range('2016-10-01 00:00:14', periods=5760*d, freq='15S')
# print time_b

conn = MySQLdb.connect("localhost","conn","hems")
c = conn.cursor()

query_empty = """truncate table check_event.discount10""" 
c.execute(query_empty)
print "check_event.discount10 are Empty!!! \n"

target = 8.42587317551 
query = "SELECT * FROM device.load_MONTH WHERE Timestamp >= '%s'   AND Timestamp <= '%s'; "% ('2016-09-03 00:00:00','2016-09-30 23:59:59')
df = pd.read_sql(query,conn)

query1 = "SELECT * FROM data_timeline.load_MONTH_1 WHERE Timestamp >= '%s'   AND Timestamp <= '%s'; "% ('2016-10-01 00:00:00','2016-10-02 23:59:59')
df1 = pd.read_sql(query1,conn)

query2 = "SELECT * FROM data_timeline.load_MONTH_1 WHERE Timestamp >= '%s'   AND Timestamp <= '%s'; "% ('2016-10-01 00:00:00','2016-10-30 23:59:59')
df2 = pd.read_sql(query2,conn)

df['watt'] = df['watt'].astype(float)
df1['watt'] = df1['watt'].astype(float)
df2['watt'] = df2['watt'].astype(float)

frames = [df, df1]
df_result = pd.concat(frames)
df_result = df_result.reset_index()
sub_power15sec = (df_result.watt*(1-(target/100)))
# print sub_power15sec.sum()

while (j <= len(df_result)-1): 
                  
    cumulasub_15sec  +=  sub_power15sec[j]        
    cumulawatt_now +=  df2.watt[j]             
    c.execute("INSERT INTO check_event.discount10 (Timestamp, target10watt,watt_now,sub_power,sub_power15sec,active5,active10,active15) VALUES (%s, %s, %s,%s, %s, %s, %s, %s)",
              (time_b[j],cumulasub_15sec/240,cumulawatt_now/240,(cumulasub_15sec-cumulawatt_now)/240,(sub_power15sec[j]-df2.watt[j])/240,0,0,0))
    conn.commit()
    j +=1 
    print ".",
print "DONE"    
```
