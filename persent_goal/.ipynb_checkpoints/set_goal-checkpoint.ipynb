```python
                                                                                                                                                                      
import pandas as pd
import MySQLdb
import time
import datetime
import sys

from sympy import *
from sympy.solvers import solve
from sympy.parsing.sympy_parser import parse_expr
```

```python
# Parametr of goal function
x = raw_input('What about the goal of electrical bill price in next month? (in bath)')
goal = float(x)

ft = raw_input('What about Ft?')
Ft = float(ft)
```

```python
# Parametr of bill function

t1 = 3.2484
t2 = 4.2218
t3 = 4.4217
ser = 38.22
u = symbols('u')

def inbill(goal):
    test_1 = solve(Eq((u*t1)+ser+(u*Ft)+(0.07*((u*t1)+ser))+(0.07*(u*Ft)),goal))
#     print test_1
    test_2 = solve(Eq((150*t1)+((u-150)*t2)+ser+(u*Ft)+(0.07*((150*t1)+((u-150)*t2)+ser))+(0.07*(u*Ft)),goal))
#     print test_2
    test_3 = solve(Eq((150*t1)+(250*t2)+((u-400)*t3)+ser+(u*Ft)+(0.07*((150*t1)+(250*t2)+((u-400)*t3)+ser))+(0.07*(u*Ft)),goal))
#     print test_3

    if 1 <=  test_1[0] <= 150:
        unit = test_1[0]
    elif 151 <=  test_2[0] <=  400:
        unit = test_2[0]
    else:
        unit = test_3[0]
    print "all unit of this month equal to",round(unit),"KW/Hr"
    return unit
    
unit =inbill(goal)
```

```python
d = 30
j = 0
sumunit = 0
time_a = pd.date_range('2016-09-01', periods=d, freq='1D')
time_b = pd.date_range('2016-09-30 23:59:59', periods=d, freq='1D')

conn= MySQLdb.connect("localhost","conn","hems")
c = conn.cursor()

query = "SELECT Timestamp,unit FROM watt.unit_per_day WHERE Timestamp >= '%s'   AND TimeStamp <= '%s'; "% (time_a[0],time_b[-1])
df = pd.read_sql(query, conn)
df['unit'] = df['unit'].astype(float)
sumunit =df.unit.sum()

target = ((sumunit-round(unit))/sumunit)*100
print "my discount target in persent equal to",target,"%" 

    
```

```python
####### set databases_watt.watt_target  to empty #########
query_empty = """truncate table watt.watt_target """ 
c.execute(query_empty)
print "watt.watt_target  are Empty!!! \n"

i =0
d= 30
time_a = pd.date_range('2016-09-01', periods=d, freq='1D')
time_b = pd.date_range('2016-09-01 23:59:59', periods=d, freq='1D')


time_b_1 = pd.date_range('2016-10-01 00:14:59', periods=96*d, freq='15T')


conn= MySQLdb.connect("localhost","conn","hems")
c = conn.cursor()

query = "SELECT Timestamp,watt FROM watt.watt_month WHERE Timestamp >= '%s'   AND TimeStamp <= '%s'; "% (time_a[0],time_b[-1])
df = pd.read_sql(query, conn)
# print query


target2 = 0
result_target = 0
while (i <= len(df)-1 ):
    result_target = (df.watt[i])*(1-(target/100))
    target2 += result_target
    c.execute("INSERT INTO watt.watt_target (Timestamp, watt,target) VALUES ( %s, %s,%s)",(time_b_1[i],result_target,target2))
    conn.commit()

    i+=1
print "DONE !!"

```
