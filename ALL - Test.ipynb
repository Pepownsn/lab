```python
import matplotlib.pyplot as plt
import pandas as pd
import MySQLdb
import numpy as np
import skfuzzy as fuzz
%pylab inline
```

# Connect to MySQL Database

* Weather forecast use `df_w_f`
- EL Senser (watt) use `df_el`
- PV (Watt , Amp and Volt) use `df_pv`
+ Temp and Humidity sensor use `df_temp`

```python
conn = MySQLdb.connect("localhost","conn","hems","power")


# EL sensor datasets  Note: 7-08-16 to 11-08-16
query_el = """
SELECT *
FROM power.watt_15T_old
WHERE TimeStamp >= "2016-08-07 00:00:00"
  AND TimeStamp < "2016-08-08 00:00:00";
"""


# Temp and Humidity sensor datasets
query_t = """
SELECT *
FROM Test.temp
WHERE TimeStamp >= "2016-10-24 00:00:00"
  AND TimeStamp < "2016-10-25 00:00:00";
"""

# PV Datasests
query_pv = """
SELECT Timestamp,V_pv,A_pv,W_pv
FROM Test.pv
WHERE Timestamp >= "2016-10-24 00:00:00"
  AND Timestamp < "2016-10-25 00:00:00";
"""

# PV avg 15T Datasests
query_pv_15t = """
SELECT Timestamp,watt
FROM PV.15T
WHERE Timestamp >= "2016-10-01 00:00:00"
  AND Timestamp < "2016-10-02 00:00:00";
"""

# Weather forecast datasets
query_weather = """
SELECT Timestamp,temp_s,hum_s,rain_1h,Sky_1
FROM Test.weather
WHERE TimeStamp >= "2016-10-24 00:00:00"
  AND TimeStamp < "2016-10-25 00:00:00";
"""

df_el = pd.read_sql(query_el, conn, index_col=['Timestamp'])

df_pv = pd.read_sql(query_pv, conn, index_col=['Timestamp'])

df_pv_15t = pd.read_sql(query_pv_15t, conn, index_col=['Timestamp'])

df_temp = pd.read_sql(query_t, conn, index_col=['Timestamp'])

df_w_f = pd.read_sql(query_weather, conn, index_col=['Timestamp'])

conn.close()

#df['temp'] = df['temp'].astype(float)
#pandas.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', 20)
```

### Plot EL Power

```python
fig, ax = plt.subplots()
ax.set_title('EL (watt)')
df_el.plot(ax=ax)
plt.show()

#df_el.to_csv('out.csv')
print "Size of Data: ", len(df_el)
```

### PV Power 15T

```python
fig, ax = plt.subplots()
ax.set_title('PV Watt/15min')
df_pv_15t.plot(ax=ax)
plt.show()

#df_el.to_csv('out.csv')
print "Power : ", sum(df_pv_15t.watt) ,"Watt/Day"
print "Size of Data: ", len(df_pv_15t)
```

### Plot Data Temp and Humidity

```python
fig, ax = plt.subplots()
ax.set_title('Temp&Hum')
df_temp.temp.plot(ax=ax)
df_temp.hum.plot(ax=ax)
ax.set_ylim(0, 110);
plt.show()

print "Size of Data: ", len(df_temp)
```
