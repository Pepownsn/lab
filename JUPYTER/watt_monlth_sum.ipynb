```python
import pandas as pd
import matplotlib.pyplot as plt
import MySQLdb
import time

j = 0
d = 20 #Day
all_p = 0
min_p = 0
sum_p = []

 # GET DATA 
conn= MySQLdb.connect("localhost","conn","hems","power")
c = conn.cursor()


####### set databases_monthly_fee to empty #########
query_empty = """truncate table watt.unit_per_day_sum""" 
c.execute(query_empty)
print "watt.unit_per_day_sum are Empty!!! \n"


# Print iterations progress
# ราย15 วินาทีเป็น 15นาที แสดงว่ามีข้อมูล 96 ค่าข้อมูลต่อหนึ่งวัน ดังนั้นถ้าต้องการข้อมูลในรายชม.ในหนึ่งวัน จะต้องได้ 24ชั่วโมง = 24 ค่าข้อมูล 
time_a = pd.date_range('2016-09-01', periods=96*d, freq='15T')
time_b = pd.date_range('2016-09-01 00:14:59', periods=96*d, freq='15T')

clock = time_a.strftime("%H:%M:%S")
clock1 = time_b.strftime("%H:%M:%S")


while  (j <= (96*d)-1):
    query = "SELECT Timestamp,s11 FROM power.enegate_bkk WHERE Timestamp >= '%s'   AND TimeStamp <= '%s'; "% (time_a[j],time_b[j])
    df = pd.read_sql(query, conn)
    df = df[df.s11 != "----"]
    df = df.reset_index(drop=True)
    df['s11'] = df['s11'].astype(float)
    
    if len(df) == 0 :
        min_p = 0
    else :
        for i in range(len(df)):
            min_p += df.s11[i]        
        min_p = min_p/len(df)/4  ##ข้อมูลพลังงานราย ชม.
    all_p += min_p
#     print time_b[j] , "-->" ,min_p/1000 ," KW/15min." ,all_p/1000 ," KW/15min."
  
    if clock1[j] == u'23:59:59':       
        c.execute("INSERT INTO watt.unit_per_day_sum(Timestamp, kwatt) VALUES ( %s, %s)",(time_b[j], all_p/1000))
        conn.commit()
        print time_b[j] , "-->"  ,all_p/1000 ," KW/15min."
    min_p = 0
    j += 1

   
    
 



    
    

     
  
```
