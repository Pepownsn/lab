```python
"""
Bar chart demo with pairs of bars grouped for easy comparison.
"""
import numpy as np
import matplotlib.pyplot as plt

query_MSE = "SELECT * FROM prediction.MSE WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2007-01-18 00:00:00","2007-01-18 11:59:59")
df = pd.read_sql(query_MSE, conn, index_col=['Timestamp'])

n_groups = len(df)
print n_groups 

MSE_holtload = df.MSE_holtload
MSE_holtsub= df.MSE_holtsub
fig, ax = plt.subplots(figsize=(12, 6))

index = np.arange(n_groups)
bar_width = 0.4

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = plt.bar(index, MSE_holtload, bar_width,
                 alpha=opacity,
                 color='b',
                 error_kw=error_config,
                 label='MSE_holtload')

rects2 = plt.bar(index + bar_width,MSE_holtsub, bar_width,
                 alpha=opacity,
                 color='r',
                 error_kw=error_config,
                 label='MSE_holtsub')
plt.xlabel('DATA SETs')
plt.ylabel('MSE')
plt.title('MSE between holtload VS holtsub')
plt.xticks(index + bar_width / 2)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
count = 0
d = 14
sumsum1 = []
sumsum2 = []
sum_1 = 0
sum_2 = 0

time_a = pd.date_range('2007-01-15 00:00:00', periods=d, freq='1D')
time_b = pd.date_range('2007-01-15 23:59:59', periods=d, freq='1D')

while (count <= d-1):
    query = "SELECT * FROM prediction.MSE WHERE Timestamp >= '%s'   AND  Timestamp <= '%s'; " % (time_a[count],time_b[count])
    df = pd.read_sql(query, conn)

    for i in range(len(df.MSE_holtload)):
        sum_1 +=  df.MSE_holtload[i]   
        sum_2 += df.MSE_holtsub[i]
    sumsum1.append(sum_1/1000000)  
    sumsum2.append(sum_2/1000000)     
    sum_1 = 0
    sum_2 = 0
    count += 1
      
print(".DONE!")        
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = len(sumsum1)
MSE_holtload = sumsum1

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars
fig, ax = plt.subplots(figsize=(14, 8))
rects1 = ax.bar(ind, MSE_holtload, width, alpha=0.4,color='r')

MSE_holtsub = sumsum2
rects2 = ax.bar(ind + width, MSE_holtsub, width,alpha=0.4, color='b')

# add some text for labels, title and axes ticks
ax.set_ylabel('% ERROR')
ax.set_title('% ERROR OF holtload VS holtsub')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('day1', 'day2', 'day3', 'day4', 'day5' , 'day6' , 'day7' , 'day8' , 'day9' , 'day10', 'day11', 'day12', 'day13', 'day14' ))
ax.legend((rects1[0], rects2[0]), ('MSE_holtload', 'MSE_holtsub'))

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2.5, 1.00*height,
                '%1.2f' % float(height),
                ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)
plt.ylim((1.5,6.25))
plt.show()
```
