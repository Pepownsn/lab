```python
import pandas as pd
import MySQLdb
import numpy as np
import time
import datetime 
import random
import matplotlib.pyplot as plt
%matplotlib inline 

conn = MySQLdb.connect("localhost","conn","hems")
c = conn.cursor()

j = 0
d = 31

time_a = pd.date_range('2016-10-01 00:00:00 ', periods=96*d, freq='15T')
time_b = pd.date_range('2016-10-01 00:14:59', periods=96*d, freq='15T')


time_a_1 = pd.date_range('2016-09-03 00:00:00 ', periods=96*30, freq='15T')
time_b_1 = pd.date_range('2016-09-03 00:14:59', periods=96*30, freq='15T')

query_empty = """truncate table PVBATT.concluded_p15min""" 
c.execute(query_empty)
print "PVBATT.concluded_p15min are Empty!!! \n"
print "Progress:",

while (j <= (96*d)-1) :   
 
    query = "SELECT Timestamp, watt FROM PVBATT.P15min_lastmonth_arranged WHERE Timestamp >= '%s'   AND TimeStamp <= '%s'; "% (time_a_1[j],time_b_1[j])
    df = pd.read_sql(query, conn)
    
    query1 = "SELECT Timestamp, watt_10,watt_15,watt_20 FROM PVBATT.target_15min WHERE Timestamp >= '%s'   AND TimeStamp <= '%s'; "% (time_a[j],time_b[j])
    df1 = pd.read_sql(query1, conn)
     
    query2 = "SELECT Timestamp, watt FROM PVBATT.P15min_observedmonth WHERE Timestamp >= '%s'   AND TimeStamp <= '%s'; "% (time_a[j],time_b[j])
    df2 = pd.read_sql(query2, conn)
    
    query3 = "SELECT Timestamp, watt FROM PVBATT.P15min_prediction WHERE Timestamp >= '%s'   AND TimeStamp <= '%s'; "% (time_a[j],time_b[j])
    df3 = pd.read_sql(query3, conn)
    
    query4 = "SELECT Timestamp, PV FROM PVBATT.pv WHERE Timestamp >= '%s'   AND TimeStamp <= '%s'; "% (time_a[j],time_b[j])
    df4 = pd.read_sql(query4, conn)
    
    new_target = df3.watt[0]*((100-20.31160051)/100)
    
    c.execute("INSERT INTO PVBATT.concluded_p15min (Timestamp, last_watt, target_20, target_15, target_10, watt,predicted_watt,new_target_20 ,PV) VALUES ( %s, %s, %s, %s, %s, %s, %s, %s, %s)",(time_b[j],df.watt[0],df1.watt_20[0],df1.watt_15[0],df1.watt_10[0],df2.watt[0],df3.watt[0],new_target,df4.PV[0]))  
    conn.commit()
    j += 1   
    print ".",
  
print(".DONE!")

```

```python
%matplotlib inline
conn = MySQLdb.connect("localhost","conn","hems")
cursor = conn.cursor()

query = "SELECT Timestamp,last_watt,watt,predicted_watt FROM PVBATT.concluded_p15min WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2016-10-01 00:00:00","2016-10-02  23:59:59")
df = pd.read_sql(query, conn )

from sklearn.metrics import mean_absolute_error,mean_squared_error
df.APE = (np.abs((df.watt[0] - df.predicted_watt[0])/df.watt[0]))*100
MAPE = np.mean(df.APE)

i = 0
MAE = []
while (i<= len(df.watt)-1):
    MAE_point = mean_absolute_error([df.watt[i]],[df.predicted_watt[i]])
    MAE.append(MAE_point)
    i+=1    
df1 = pd.DataFrame(MAE)
cols = ['error']
df1.columns = cols
print max(df1.error)
# df1['ERROR'] = df1

print "% MEAN ABSOLUTE PERCENT ERROR :",MAPE, "%"
print "Size of Data_excatly_load: ", len(df.watt) , " Day =", len(df.watt)/96
print "Size of Data_pred_load   : ", len(df.predicted_watt) , " Day =", len(df.predicted_watt)/96


fig, (ax0, ax1) = plt.subplots(nrows=2,figsize=(9*2, 6*2))
ax0.set_title('SECOND CRITERION')
ax1.set_title('Errors of prediction')

# ax0.set_title('Data in 15 Minute @ 10W')
df.plot(ax=ax0)
df1.plot(ax=ax1)

plt.show()
```
