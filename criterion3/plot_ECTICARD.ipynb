```python
"""
Demo of table function to display a table within a plot.
"""
import numpy as np
import matplotlib.pyplot as plt


data = [[ 287.7209, 302.2008, 288.2051, 308.5747, 290.2195, 284.0863, 280.2423, 210.2412],
        [  79.4541, 45.3694, 23.8447, 0.0, 18.4067, 25.2892, 30.6247, 157.0338],
        [  8.9647, 0.00 , 0.00, 0.00, 3.1180, 17.7602, 55.4742, 60.01405]]


columns = ('PV1200&BATT3000', 'PV600&BATT3000' ,'PV300&BATT3000','No-PV&No-BATT', 'PV300&No-BATT', 'PV600&No-BATT', 'PV1200&No-BATT', 'PV3000&BATT6000')
rows = ('GRID_Consumption (KW)', 'PV_BATT_Consumption (KW)',' Over_PV_power (KW)')
plt.subplots(figsize=(16,6))

values = np.arange(0, 520, 20)
# value_increment = 0.5
colors = ('#ffb5d5','#baeaef','#efe8ba')
n_rows = len(data)
index = np.arange(len(columns)) +0.25
bar_width = 0.5

y_offset = np.array([0.0] * len(columns))
cell_text = []

for row in range(n_rows):
#     print index, data[row], bar_width,y_offset, colors[row]
    plt.bar(index, data[row], bar_width, bottom=y_offset, color=colors[row])
    y_offset= y_offset+data[row]
    cell_text.append(['%1.4f' % (x) for x in data[row]])

the_table = plt.table(cellText=cell_text, 
                      rowLabels=rows,
                      rowColours=colors,
                      colLabels=columns, fontsize=12,
                      cellLoc='center',                      
                      loc='bottom',)
the_table.auto_set_font_size(False)
the_table.set_fontsize(13.5)
the_table.scale(1, 2.5)
plt.subplots_adjust(left=0, bottom=0)

# plt.yticks(values * value_increment, ['%d' % val for val in values])
plt.xticks([])
plt.ylabel('Power(watt)', fontsize=14)
plt.title('Power Consumption by PV and BATTERY and GRID', fontsize=16)
plt.show()
```

```python
# a stacked bar plot with errorbars
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
N = 7
GRID_Consumption = (287.7209, 302.2008, 288.2051, 308.5747, 290.2195, 284.0863, 280.2423)
PV_BATT_Consumption  = ( 79.4541, 45.3694, 23.8447, 0.0, 18.4067, 25.2892, 30.6247)
Over_PV_power = (8.9647, 0.00 , 0.00, 0.00, 3.1180, 17.7602, 55.4742)

ind = np.arange(N)    # the x locations for the groups
width = 0.4       # the width of the bars: can also be len(x) sequence
plt.subplots(figsize=(14,6))

p1 = plt.bar(ind+0.4, GRID_Consumption, width, color='#d7d6d8')
p2 = plt.bar(ind+0.4, PV_BATT_Consumption, width,bottom=GRID_Consumption, color='#c4b8d1')
p3 = plt.bar(ind+0.4, Over_PV_power, width,bottom=[i+j for i,j in zip(GRID_Consumption,PV_BATT_Consumption)], color='#806999')

plt.ylabel('Power(watt)', fontsize=14)
plt.title('Power Consumption by PV and BATTERY and GRID', fontsize=14)
plt.xticks(ind+0.4+0.2, ('PV1200&BATT3000', 'PV600&BATT3000' ,'PV300&BATT3000','No-PV&No-BATT', 'PV300&No-BATT', 'PV600&No-BATT', 'PV1200&No-BATT', 'PV3000&BATT6000'),rotation='vertical')
plt.yticks(np.arange(0,550,20))
plt.legend((p1[0], p2[0], p3[0]), ('GRID_Consumption(KW)', 'PV_BATT_Consumption(KW)', 'Over_PV_power (KW)'))
plt.show()
```

```python
"""
Bar chart demo with pairs of bars grouped for easy comparison.
"""
import numpy as np
import matplotlib.pyplot as plt


n_groups = 8

GRID_Consumption = (287.7209, 302.2008, 288.2051, 308.5747, 290.2195, 284.0863, 280.2423, 210.2412)
PV_BATT_Consumption  = (79.4541, 45.3694, 23.8447, 0.0, 18.4067, 25.2892, 30.6247, 157.0338)
total_power = [i+j for i,j in zip(GRID_Consumption,PV_BATT_Consumption)]
print total_power 
price_bath = (1169.62, 1234.28, 1171.78, 1262.75, 1180.78, 1153.38, 1136.22, 823.59)
theaffect = (0,907,2646,2866,2848,2808,2699,0)

plt.subplots(figsize=(14,6))
# fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.2

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = plt.bar(index+bar_width, total_power, bar_width,
                 alpha=opacity,
                 color='#d7d6d8',
                 error_kw=error_config,
                 label='totalpower (KW)')

rects2 = plt.bar(index + bar_width+bar_width, price_bath, bar_width,
                 alpha=opacity,
                 color='#c4b8d1',
                 error_kw=error_config,
                 label='price (BAHT)')

rects3 = plt.bar(index + bar_width+bar_width+bar_width, theaffect, bar_width,
                 alpha=opacity,
                 color='#806999',
                 error_kw=error_config,
                 label="the number of  user's affect (TIMES)")

plt.ylabel('Price(baht)')
plt.title('Total Power Consumption ')
plt.yticks(np.arange(0, 4000, 500))
plt.xticks(index + bar_width+ bar_width+bar_width / 2, ('PV1200&BATT3000', 'PV600&BATT3000' ,'PV300&BATT3000','No-PV&No-BATT', 'PV300&No-BATT', 'PV600&No-BATT', 'PV1200&No-BATT', 'PV3000&BATT6000'))
plt.legend()
plt.tight_layout()

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1*height,
                '%d' % int(height),
                ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.show()
```
