```python
import pandas as pd
import MySQLdb
import numpy as np
import time
import datetime 
import random
import matplotlib.pyplot as plt
%matplotlib inline 

conn = MySQLdb.connect("localhost","conn","hems")
c = conn.cursor()
min_p = float(0)
j = 0
d = 31
time_a = pd.date_range('2016-10-01 06:45:00 ', periods=96*d, freq='15T')
# print time_a
time_b = pd.date_range('2016-10-01 06:59:59', periods=96*d, freq='15T')
# print time_b

# query_empty = """truncate table PVBATT.P15min_observedmonth""" 
# c.execute(query_empty)
# print "PVBATT.P15min_observedmonth are Empty!!! \n"
# print "Progress:",

while (j <= (96*d)-1) :   
    query = "SELECT Timestamp, watt FROM data_timeline.load_MONTH_1 WHERE Timestamp >= '%s'   AND TimeStamp <= '%s'; "% (time_a[j],time_b[j])
    df = pd.read_sql(query, conn)
    df.cTime =  pd.Series(pd.to_datetime(df.Timestamp))
    df['watt'] = df['watt'].astype(float)
    
    if len(df) == 0 :
        min_p = 0
    else :
        for i in range(len(df.cTime.dt.minute)):
            min_p += df.watt[i]        
        min_p = min_p/len(df)/4  
        
        #df.to_csv('out.csv')
    print time_b[j] , "-->" ,min_p ," watt/15min"
    c.execute("INSERT INTO PVBATT.P15min_observedmonth (Timestamp, watt) VALUES ( %s, %s)",(time_b[j],min_p))  
    
    #print "|",
    conn.commit()
    min_p = 0
    j += 1   
    print ".",
    
print time_a[j],time_b[j]
print(".DONE!")
```
