```python
import matplotlib.pyplot as plt
import pandas as pd # v. 1.7
import MySQLdb
import numpy as np
import time
import sys
import datetime 
from datetime import datetime
from sympy import *
from sympy.solvers import solve
from sympy.parsing.sympy_parser import parse_expr
from sklearn.metrics import mean_absolute_error,mean_squared_error
import statistics 
%matplotlib inline


conn = MySQLdb.connect("localhost","conn","hems")
cursor = conn.cursor()

query_totalpower = "SELECT * FROM prediction.pred_moving_totalpower WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2007-02-17 00:00:00","2007-02-20 23:59:59")
df_totalpower = pd.read_sql(query_totalpower, conn , index_col=['Timestamp'])

query = "SELECT Timestamp, active_power FROM prediction.data15min_4y WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2007-02-17 00:00:00","2007-02-20 23:59:59")
df = pd.read_sql(query, conn , index_col=['Timestamp'])

# df['holtsub'] = df_holt_sub.watt_sumsub
df['pred_moving_totalpower'] = df_totalpower.watt
fig, ax = plt.subplots(figsize=(14, 6))
ax.set_title('exactly_power VS pred_moving_totalpower',  fontsize=12)
plt.xlabel('Timestamp', fontsize=12)
plt.ylabel('watt', fontsize=12)
df.plot(ax=ax)
plt.show()


from sklearn.metrics import mean_absolute_error,mean_squared_error
df.APE = (np.abs((df.active_power-df_totalpower.watt)/df.active_power))*100
MAPE = np.mean(df.APE)
# df.to_csv('out.csv')

print "% MEAN ABSOLUTE PERCENT ERROR :",MAPE, "%"
print "Size of Data_excatly_load: ", len(df) , " Day =", len(df)/96
print "Size of Data_pred_load   : ", len(df_totalpower) , " Day =", len(df_totalpower)/96
```

```python
from sklearn.metrics import mean_absolute_error



conn = MySQLdb.connect("localhost","conn","hems")
c= conn.cursor()

query_empty = """truncate table prediction.MAE""" 
c.execute(query_empty)
print "prediction.MAE are Empty!!! \n"


query = "SELECT Timestamp, active_power FROM prediction.data15min_4y WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2007-02-01 00:14:59","2007-02-01 23:59:59")
df = pd.read_sql(query, conn , index_col=['Timestamp'])

query_totalpower = "SELECT * FROM prediction.pred_moving_totalpower WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2007-02-01 00:14:59","2007-02-01 23:59:59")
df_moving_totalpower = pd.read_sql(query_totalpower, conn , index_col=['Timestamp'])

query_mtall = "SELECT * FROM prediction.pred_moving_mtall WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2007-02-01 00:14:59","2007-02-01 23:59:59")
df_moving_mtall = pd.read_sql(query_mtall, conn , index_col=['Timestamp'])

query_totalpower = "SELECT * FROM prediction.pred_expand_totalpower WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2007-02-01 00:14:59","2007-02-01 23:59:59")
df_expanding_totalpower = pd.read_sql(query_totalpower, conn , index_col=['Timestamp'])

query_mtall = "SELECT * FROM prediction.pred_expand_mtall WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2007-02-01 00:14:59","2007-02-01 23:59:59")
df_expanding_mtall = pd.read_sql(query_mtall, conn , index_col=['Timestamp'])


time_a = pd.date_range('2007-02-01 00:14:59', periods=(96*1)+1, freq='15T')
i = 0
while (i<= len(df)-1):
    y_true = df.active_power[i]
    y_moving_totalpower  = df_moving_totalpower.watt[i]
    y_moving_mtall  = df_moving_mtall.watt_sumsub[i]
    y_expanding_totalpower  = df_expanding_totalpower.watt[i]
    y_expanding_mtall  = df_expanding_mtall.watt_sumsub[i]

    MAE_moving_totalpower = mean_absolute_error([y_true], [y_moving_totalpower])
    MAE_moving_mtall = mean_absolute_error([y_true],[y_moving_mtall])
    MAE_expanding_totalpower = mean_absolute_error([y_true],[y_expanding_totalpower])
    MAE_expanding_mtall = mean_absolute_error([y_true],[y_expanding_mtall])
    
    c.execute("INSERT INTO prediction.MAE (Timestamp,MAE_moving_totalpower,MAE_moving_mtall,MAE_expanding_totalpower,MAE_expanding_mtall) VALUES ( %s, %s, %s, %s, %s)",(time_a[i],MAE_moving_totalpower,MAE_moving_mtall ,MAE_expanding_totalpower ,MAE_expanding_mtall))
    conn.commit()
    i+=1
           

print "!!!!!!!!   DONE    !!!!!!!!!"
```
