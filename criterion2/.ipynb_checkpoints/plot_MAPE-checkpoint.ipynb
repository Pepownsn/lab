```python
"""
Bar chart demo with pairs of bars grouped for easy comparison.
"""
import numpy as np
import matplotlib.pyplot as plt

query_MAE = "SELECT * FROM prediction.MAE WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2007-02-15 00:00:00","2007-02-15 12:59:59")
df = pd.read_sql(query_MAE, conn, index_col=['Timestamp'])

n_groups = len(df)
print n_groups 

MAE_totalpower= df.MAE_totalpower
MAE_mtall= df.MAE_mtall
fig, ax = plt.subplots(figsize=(12, 6))
index = np.arange(n_groups)
bar_width = 0.4

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = plt.bar(index, MAE_totalpower, bar_width,
                 alpha=opacity,
                 color='b',
                 error_kw=error_config,
                 label='MAE_moving_totalpower')

rects2 = plt.bar(index + bar_width,MAE_mtall, bar_width,
                 alpha=opacity,
                 color='r',
                 error_kw=error_config,
                 label='MAE_moving_mtall')
plt.xlabel('DATA SETs')
plt.ylabel('MAE')
plt.title('MAE between MAE_moving_totalpower VS MAE_moving_mtall')
plt.xticks(index + bar_width / 2)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
count = 0
d = 14
sumsum1 = []
sumsum2 = []
sum_1 = 0
sum_2 = 0

time_a = pd.date_range('2007-02-15 00:00:00', periods=d, freq='1D')
time_b = pd.date_range('2007-02-15 23:59:59', periods=d, freq='1D')

while (count <= d-1):
    query = "SELECT * FROM prediction.MAE WHERE Timestamp >= '%s'   AND  Timestamp <= '%s'; " % (time_a[count],time_b[count])
    df = pd.read_sql(query, conn)

    for i in range(len(df.MAE_mtall)):
        sum_1 +=  df.MAE_totalpower[i]   
        sum_2 += df.MAE_mtall[i]
    sumsum1.append(sum_1/1000000)  
    sumsum2.append(sum_2/1000000)     
    sum_1 = 0
    sum_2 = 0
    count += 1
      
print(".DONE!")        
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = len(sumsum1)
MAE_totalpower = sumsum1

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars
fig, ax = plt.subplots(figsize=(14, 8))
rects1 = ax.bar(ind, MAE_totalpower, width, alpha=0.4,color='r')

MAE_mtall = sumsum2
rects2 = ax.bar(ind + width, MAE_mtall, width,alpha=0.4, color='b')

# add some text for labels, title and axes ticks
ax.set_ylabel('MAE')
ax.set_title(' MAE_moving_totalpower VS MAE_moving_mtall')
ax.set_xticks(ind + width / 2)
#ax.set_xticklabels(('day1', 'day2', 'day3', 'day4', 'day5' , 'day6' , 'day7' , 'day8' , 'day9' , 'day10', 'day11', 'day12', 'day13', 'day14' ))
ax.legend((rects1[0], rects2[0]), ('MAE_moving_totalpower', 'MAE_moving_mtall'))

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2.5, 1.00*height,
                '%1.2f' % float(height),
                ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)
#plt.ylim((1.5,6.25))
plt.show()
```
