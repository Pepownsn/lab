```python
import matplotlib.pyplot as plt
import pandas as pd # v. 1.7
import MySQLdb
import numpy as np
import time
import sys
import datetime 
from datetime import datetime
from sympy import *
from sympy.solvers import solve
from sympy.parsing.sympy_parser import parse_expr
from sklearn.metrics import mean_absolute_error,mean_squared_error
import statistics 
%matplotlib inline


conn = MySQLdb.connect("localhost","conn","hems")
cursor = conn.cursor()


query_totalpower = "SELECT Timestamp,watt_sub1 FROM prediction.pred_moving_mtall WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2007-02-15 00:00:00","2007-02-20 23:59:59")
df_totalpower = pd.read_sql(query_totalpower, conn , index_col=['Timestamp'])

query = "SELECT Timestamp,Sub_metering_1 FROM prediction.data15min_4y WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2007-02-15 00:00:00","2007-02-20 23:59:59")
df = pd.read_sql(query, conn , index_col=['Timestamp'])

# df['holtsub'] = df_holt_sub.watt_sumsub
df['pred_moving_totalpower_meter1'] = df_totalpower.watt_sub1 
fig, ax = plt.subplots(figsize=(14, 6))
ax.set_title('exactly_power VS pred_moving_totalpower at sub-meter1')
df.plot(ax=ax)
plt.show()
```
